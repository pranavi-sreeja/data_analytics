Description of the Dataset

Each row represents a student’s exam performance.

Some values are missing (NaN), which indicates that some students may not have attempted certain questions.

Marks range within the limits specified by the column names (e.g., "Q5M10" has a max of 10).

The Total column aggregates marks from all attempted questions.

Overview of the Dataset
The dataset appears to contain marks for different questions in an exam. The columns represent:

Total: The total marks obtained by a student.

Q1aM4, Q1bM6, ..., Q6bM6: Marks obtained in different parts of the exam, where:

"Q1aM4" likely means "Question 1, Part a, Maximum 4 Marks".

"Q1bM6" means "Question 1, Part b, Maximum 6 Marks", and so on.

The numbers at the end indicate the maximum possible marks for that question.

This suggests either optional questions or students skipping certain parts.
