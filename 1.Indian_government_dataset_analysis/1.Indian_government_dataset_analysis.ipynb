```python
from scipy.stats import ttest_ind
import pandas as pd
pm10 = df.loc[df['pollutant_id'] == 'PM10', 'pollutant_avg'].dropna()
no2 = df.loc[df['pollutant_id'] == 'NO2', 'pollutant_avg'].dropna()
t_stat, p_value = ttest_ind(pm10, no2)
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
if p_value < 0.05:
    print("There is a statistically significant difference between PM10 and NO2 levels.")
else:
    print("There is no statistically significant difference between PM10 and NO2 levels.")
```

REPORT ON DATASET
AIR QUALITY INDEX ANALYSIS REPORT
 1. INTRODUCTION
 This report presents an analysis of air-quality index across different states and cities in INDIA. The dataset is cleaned, visualized, and subjected to statistical analysis, including t-tests and chi-square tests.The goal of this analysis is to identify patterns in air quality, assess pollutant levels, and derive actionable insights. 

 2. DATA PRE-PROCESSING
 Checked for missing values and filled them appropriately.Categorical data such as country, state, city, and pollutant_id were filled using mode values. Numerical values were filled using the median to reduce the effect of outliers.

3. DATA VISUALIZATION
 The data visualization revealed several key insights into pollution levels across different cities and pollutants. The pie chart for average pollution levels by the top 10 cities highlighted that metropolitan areas like Delhi, Mumbai, and Hyderabad contribute significantly to overall pollution due to high population density and vehicular emissions. The individual pie charts for pollutant distribution showed that PM10 and NO2 are dominant in cities with heavy traffic and industrial activity, while higher NH3 levels were found in agricultural and waste-processing zones. A bar plot of average pollution levels by city confirmed that urban areas face consistently high pollution, while smaller cities exhibited more variation, suggesting localized pollution sources. 

 STATISTICAL ANALYSIS
 T-TEST
 The T-test was conducted to compare the average levels of PM10 and NO2 across different cities. 
 The resulting t-statistic was 28.39 with a p-value of 4.35 × 10⁻¹²⁷, which is significantly lower than the standard significance level of 0.05. 
 This indicates that there is a statistically significant difference between the average levels of PM10 and NO2, suggesting that the sources or patterns of these pollutants are different.

 CHI-SQUARE TEST
 The Chi-Square Test was used to evaluate the relationship between the type of pollutant and the city. 
 The chi-square statistic was 89.68 with a p-value of 1.0, which is greater than the significance threshold of 0.05. 
 This indicates that there is no statistically significant relationship between the type of pollutant and the city, suggesting that pollutant levels are influenced more by external environmental factors (e.g., traffic, industrial activity, weather conditions) rather than the city itself. 

 CONCLUSION
 The analysis of the air quality dataset revealed key insights into pollution patterns across different cities in India.
 After cleaning and organizing the data, it was clear that big cities like Delhi, Mumbai, and Hyderabad have higher pollution levels because of traffic and factories. Pollution levels were higher in winter due to more emissions and less air movement, while they dropped during the rainy season as rain helped clear the air.
 The analysis suggests that pollution in big cities is mainly due to traffic and industrial activities. To improve air quality, stricter pollution control measures, better traffic management, and cleaner industrial practices are needed.
